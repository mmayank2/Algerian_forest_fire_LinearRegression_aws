```python
def multicollinearity(data,threshold):
    col_corr=set()
    corr_metrix=data.corr()
    for i in range(len(corr_metrix)):
        for j in range(i):
            if abs(corr_metrix.iloc[i][j])>threshold:
                col_corr.add(corr_metrix.columns[i])
    return col_corr
```
